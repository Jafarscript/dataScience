```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
link = 'https://raw.githubusercontent.com/microsoft/Data-Science-For-Beginners/main/data/birds.csv'
birds = pd.read_csv(link)
birds.head()
```

```python
newTable = birds.groupby('Name').sum()
```

```python
wingspan = birds['MaxWingspan'] 
wingspan.plot()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.xticks(rotation=45)
x = birds['Name'] 
y = birds['MaxWingspan']

plt.plot(x, y)

plt.show()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x, y, 'bo')
    if birds['MaxWingspan'][i] > 500:
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=12)
    
plt.show()
```

```python
newTable.loc['Bald eagle']
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    if birds['Name'][i] not in ['Bald eagle', 'Prairie falcon']:
        plt.plot(x, y, 'bo')
plt.show()
```

```python
birds.plot(x='Category',
        kind='bar',
        stacked=True,
        title='Birds of Minnesota')
```

```python
category_count = birds.value_counts(birds['Category'].values, sort=True)
plt.rcParams['figure.figsize'] = [6, 12]
category_count.plot.barh()
```

```python
maxlength = birds['MaxLength']
plt.barh(y=birds['Category'], width=maxlength)
plt.rcParams['figure.figsize'] = [6, 12]
plt.show()
```

```python
minLength = birds['MinLength']
maxLength = birds['MaxLength']
category = birds['Category']

plt.barh(category, maxLength)
plt.barh(category, minLength)

plt.show()
```

```python
filtered = birds.dropna(axis = 0)
```

```python
filtered.columns
```

```python
y = filtered.Name
```

```python
featured =['ScientificName', 'Category', 'Order','MinLength', 'MaxLength']

X = filtered[featured]

X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
bird_model = DecisionTreeRegressor(random_state = 1)
```

```python
bird_model.fit(X,y)
```

```python
#Overfitting: capturing spurious patterns that won't recur in the future, leading to less accurate predictions, or
#Underfitting: failing to capture relevant patterns, again leading to less accurate predictions.
```
